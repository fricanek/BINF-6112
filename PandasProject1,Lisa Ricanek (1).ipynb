```python
#getting the first day of the month March-Aug
March_Aug = confirmed.iloc[:,[6,50,81,111,142,172,203]]
```

```python
#finding the percent of covid cases for each month
March_Aug['March percent'] = (March_Aug['3/1/20'] / 
                      March_Aug['3/1/20'].sum()) * 100
March_Aug['April percent'] = (March_Aug['4/1/20'] / 
                      March_Aug['4/1/20'].sum()) * 100
March_Aug['May percent'] = (March_Aug['5/1/20'] / 
                      March_Aug['5/1/20'].sum()) * 100
March_Aug['June percent'] = (March_Aug['6/1/20'] / 
                      March_Aug['6/1/20'].sum()) * 100
March_Aug['July percent'] = (March_Aug['7/1/20'] / 
                      March_Aug['7/1/20'].sum()) * 100
March_Aug['August percent'] = (March_Aug['8/1/20'] / 
                      March_Aug['8/1/20'].sum()) * 100

March_Aug
```

```python
#finding the portions of cases that happened in NY
NYpercent = March_Aug[March_Aug["Province_State"]=="New York"]
```

```python
plot = NYpercent[['March percent','April percent','May percent', 'June percent', 'July percent', 'August percent']].sum().plot.bar()
plot.figure.suptitle("Proportions of confirmed cases in New York")
plot.set_xlabel("Month")
plot.set_ylabel("Percentage")
```
